```python
import numpy as np
```

```python
a = np.linspace(0,5,10)
```

```python
b = a ** 3
```

```python
a
```

```python
b
```

Функциональный метод создания графиков

```python
plt.plot(a,b) # a для оси x и для оси y
plt.xlabel('a array')
plt.ylabel('b array')
plt.title('Plot for a and b arrays')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r') # r красный цвет
plt.subplot(1,2,2)
plt.plot(b,a,'b')
```

ООП метод

```python
figure = plt.figure()
axes = figure.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b)
axes.set_xlabel('a array values')
axes.set_ylabel('b array values')
axes.set_title('Some title')
```

```python
figure = plt.figure()
axes1 = figure.add_axes([0.1,0.1,0.8,0.8]) #процентное соотношение
axes2 = figure.add_axes([0.3,0.6,0.5,0.4]) #процентное соотношение 03-слево, 0.6 - снизу, 0.5-ширина 0.4-высота

```

```python
figure = plt.figure()
axes1 = figure.add_axes([0,0,1,1]) #процентное соотношение
axes2 = figure.add_axes([0.1,0.4,0.5,0.4]) 

axes1.plot(a,b)
axes2.plot(b,a)
axes1.set_xlabel('a array values')
axes1.set_ylabel('b array values')
axes1.set_title('Axes1 title')
axes2.set_xlabel('a array values')
axes2.set_ylabel('b array values')
axes2.set_title('Axes2 title')
```
