#### Функциональный метод создания графиков

```python
plt.plot(a,b) # a для оси x и для оси y
plt.xlabel('a array')
plt.ylabel('b array')
plt.title('Plot for a and b arrays')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r') # r красный цвет
plt.subplot(1,2,2)
plt.plot(b,a,'b')
```

ООП метод

```python
figure = plt.figure()
axes = figure.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b)
axes.set_xlabel('a array values')
axes.set_ylabel('b array values')
axes.set_title('Some title')
```

```python
figure = plt.figure()
axes1 = figure.add_axes([0.1,0.1,0.8,0.8]) #процентное соотношение
axes2 = figure.add_axes([0.3,0.6,0.5,0.4]) #процентное соотношение 03-слево, 0.6 - снизу, 0.5-ширина 0.4-высота

```

```python
figure = plt.figure()
axes1 = figure.add_axes([0,0,1,1]) #процентное соотношение
axes2 = figure.add_axes([0.1,0.4,0.5,0.4]) 

axes1.plot(a,b)
axes2.plot(b,a)
axes1.set_xlabel('a array values')
axes1.set_ylabel('b array values')
axes1.set_title('Axes1 title')
axes2.set_xlabel('a array values')
axes2.set_ylabel('b array values')
axes2.set_title('Axes2 title')
```

# Mmatplotlib. Part 2

```python
figure, axes = plt.subplots(nrows=2, ncols=2)
plt.tight_layout() # для красивого отображения
#axes.plot(a,b)
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
for current_axe in axes:
    current_axe.plot(a,b)
```

```python
axes
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(a,b)
axes[0].set_xlabel('a array values')
axes[0].set_ylabel('b array values')
axes[0].set_title('Axes1 title')

axes[1].set_xlabel('b array values')
axes[1].set_ylabel('a array values')
axes[1].set_title('Axes2 title')
axes[1].plot(b,a)
plt.tight_layout() # решение проблем с наложением графиков
```

# Figure size, aspect ratio and dpi

```python
figure = plt.figure(figsize=(4,3),dpi=200)
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
figure, axes = plt.subplots(nrows=2,ncols=1,figsize=(7,2),dpi=200)
axes[0].plot(a,b)
plt.tight_layout()
```

## Maptplotlib. Part 3

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b,color = '#11e452',linewidth=5,alpha=0.5, linestyle='-.',marker="o",markersize=20,markerfacecolor='red', markeredgewidth=3) # есть стандартные цвета,red green и т.д., но можно и 16-ричное представление
#lineweidth - ширина линии, можно использовать lw
#alpha - прозрачность
#linestyle - Стиль линии,можно короче ls @see https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
#marker - маркеры которые указывают точки на графике
#markersize- размер маркера
#markerfacecolor цвет маркера
#markeredgewidth тол
```
